```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import sklearn.linear_model
```

```python
X= np.array([1,2,3,4])
N=X.shape[0]
X = np.vstack((np.ones(N), X)).T  #or, we could do :  X.reshape((N,1))  ## then you need to change the code for plots
wGT = np.array([2, 0.1])
y = X@wGT
```

```python
plt.scatter(X[:,1],y)
plt.ylim([0,3])
```

### Question 1
Plot the Lasso Loss (as a function of the models'for $\alpha=1$ and alpha=0.01. Ignore the bias at first, setting it to its true value.

```python
w = np.array([2, 0.1])
alpha=0.1
np.sum((X@w-y)**2) + alpha*np.sum(np.abs(w))
```

```python
def LossLasso(w,X,T,alpha):
    return  np.sum((X@w-y)**2) + alpha*np.sum(np.abs(w))
    
```

```python
np.linspace(0,1,101)
```

```python
alpha=0
ws = np.linspace(-1,1,1001)
Ls = []
b=2
for w1 in ws:
    w = np.array([b,w1])
    Ls.append( LossLasso(w,X,y,alpha) )
plt.plot(ws, Ls) #, marker='+')

alpha=10
ws = np.linspace(-1,1,1000)
Ls = []
b=2
for w1 in ws:
    w = np.array([b,w1])
    Ls.append( LossLasso(w,X,y,alpha) )
plt.plot(ws, Ls) #, marker='+')
```

### Question 2
Plot the Ridge Loss for $\lambda=1$ and lambda=0.01. Ignore the bias at first, setting it to its true value.

```python
def LossRidge(w,X,T,alpha):
    return  np.sum((X@w-y)**2) + alpha*np.sum(np.abs(w)**2)
alpha=0
ws = np.linspace(-1,1,1000)
Ls = []
b=2
for w1 in ws:
    w = np.array([b,w1])
    Ls.append( LossRidge(w,X,y,alpha) )
plt.plot(ws, Ls) #, marker='+')

alpha=1
ws = np.linspace(-1,1,1000)
Ls = []
b=2
for w1 in ws:
    w = np.array([b,w1])
    Ls.append( LossRidge(w,X,y,alpha) )
plt.plot(ws, Ls) #, marker='+')
```

### Question 3
Learn the Lasso model, using the method  `sklearn.linear_model.Lasso`, and compare the results using alpha=1 or alpha=0.01.

```python
#?sklearn.linear_model.Lasso
```

```python
alpha=0.01
lasso1 = sklearn.linear_model.Lasso(alpha=alpha)
```

```python
lasso1.fit(X,y)
```

```python
prediction = lasso1.predict(X)
```

```python
plt.scatter(X[:,1],y, label="data", c='k')
plt.plot(X[:,1], prediction, ls="--", label="model, alpha=%.e"%alpha)
plt.ylim([0,3])
plt.legend()
```

```python
alpha=1
lasso1 = sklearn.linear_model.Lasso(alpha=alpha)
lasso1.fit(X,y)
prediction = lasso1.predict(X)
plt.scatter(X[:,1],y, label="data", c='k')
plt.plot(X[:,1], prediction, ls="--", label="model, alpha=%.e"%alpha)
plt.ylim([0,3])
plt.legend()
```

### Question 4:
What is the limit alpha value for which we get a 0 slope (in our model) ?

```python
alphaRange = [10**k for k in range(-6,2)]
plt.scatter(X[:,1],y, label="data", c='k')
for alpha in alphaRange:
    lasso1.set_params(alpha=alpha)
    lasso1.fit(X,y)
    print(alpha, lasso1.coef_, lasso1.intercept_)
    plt.plot(X[:,1], X@ lasso1.coef_+ lasso1.intercept_, ls="--", label="model, alpha=%.e"%alpha)

# plt.ylim([0,3])
plt.legend()
```
